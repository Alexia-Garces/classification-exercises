### Attendance Data

#### 1.  Load the attendance.csv file and calculate an attendnace percentage for each student. One half day is worth 50% of a full day, and 10 tardies is equal to one absence.

```python
attendance = pd.read_csv("attendance.csv")
attendance.head()
```

```python
#rename unamed column
attendance = attendance.rename(columns={"Unnamed: 0": "name"})
attendance.head()
```

```python
#melt dataframe
attendance_melt = attendance.melt(id_vars=["name"])
attendance_melt.head()
```

```python
#create function to assign value to attendance letter grade
# P = 100, T = .9, H = .5

def attendance_number_grade(x):
    if x == "P":
        return 100
    elif x == "T":
        return 0.9
    elif x == "H":
        return 0.5
    else:
        return 0
```

```python
#use .apply to create attendance_number_grade_column
attendance_melt['attendance_number_grade'] = attendance_melt['value'].apply(attendance_number_grade)
```

```python
#group by name to get the average number grade for attendance
attendance_melt.groupby('name').attendance_number_grade.mean()
```
